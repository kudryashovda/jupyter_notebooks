```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame({"Time": ['00:01:00', '00:02:00', '00:05:00', '00:15:00'],
                   "Price": [77,76,68,70]})
df["Time"] = pd.to_datetime(df["Time"], format = "%H:%M:%S")

sns.scatterplot(x = 'Time', y = 'Price',data = df)
```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame([['2014-01-01 00:00:01.0',10],['2014-01-01 00:00:02.0',15],['2014-01-01 00:00:06.0',15]], columns=['timestamp','Task_Length'])

df["timestamp"] = pd.to_datetime(df["timestamp"], format = "%Y-%m-%d %H:%M:%S.%f")

sns.scatterplot(x='timestamp', y='Task_Length', data=df)
```
