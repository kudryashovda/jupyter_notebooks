```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5]
y = [5,6,7,8,9,10]

df = pd.DataFrame(
    {'x': x,
     'y': y,
    })

sns.set()

#define plotting region (1 rows, 2 columns) with margins
fig, axes = plt.subplots(1, 2, constrained_layout=True)

#create boxplot in each subplot
sns.scatterplot(data=df, x='x', y='x', ax=axes[0])
sns.scatterplot(data=df, x='x', y='y', ax=axes[1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

times = ['00:00:50', '00:01:55', '00:02:18', '00:05:24', '00:05:34', '00:07:33']
descriptions = ['STATS', 'STATS', 'ONLINE', 'STATS', 'STATS', 'OFFLINE']

df = pd.DataFrame(
    {'time': times,
     'description': descriptions
    })

df['time'] = pd.to_datetime(df['time'], format="%H:%M:%S")

ax = sns.scatterplot(x='time', y='description', data=df)

myFmt = mdates.DateFormatter('%H:%M:%S')
ax.xaxis.set_major_formatter(myFmt)

ax.figure.autofmt_xdate()
plt.show()
```
